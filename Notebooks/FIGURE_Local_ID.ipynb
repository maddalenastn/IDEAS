## import

```python
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## functions

```python
def scatter_plot(x, y, bins, labelx, labely):

    plt.figure(figsize=(7, 6))

    num_bins = bins
    bins = np.linspace(min(x), max(x), num_bins + 1)
    bin_indices = np.digitize(x, bins)
    bin_means = [y[bin_indices == i].mean() for i in range(1, len(bins))]
    bin_stds = [y[bin_indices == i].std() for i in range(1, len(bins))]
    bin_centers = (bins[:-1] + bins[1:]) / 2

    plt.scatter(x, y, alpha=0.3)
    plt.scatter(bin_centers, bin_means, color="black", marker="o", s=100)
    plt.errorbar(bin_centers, bin_means, yerr=bin_stds, fmt="o", color="black", capsize=5)

    plt.xlabel(labelx, fontsize=16)
    plt.ylabel(labely, fontsize=16)
    plt.yticks(fontsize=14)
    plt.grid(color='grey', linestyle='--', linewidth=0.7, alpha=0.7)  

    sns.despine()
    plt.show()

    
```
